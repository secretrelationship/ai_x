```python
import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/124.0.0.0 Safari/537.36"
    )
}

url = "https://www.yes24.com/product/category/bestseller?categoryNumber=001&pageNumber=1&pageSize=24"
response = requests.get(url, headers=headers)
response.raise_for_status()
print(response.text)

soup = BeautifulSoup(response.text, 'html.parser')
books = soup.select('ol#categoryList > li')

for rank, book in enumerate(books, start=1):
    title = book.select_one('p.prod_name a').text.strip() if book.select_one('p.prod_name a') else '제목없음'
    author = book.select_one('span.prod_author').text.strip() if book.select_one('span.prod_author') else '저자없음'
    
    print(f'{rank}위 | 제목: {title} | 저자: {author}')

```
